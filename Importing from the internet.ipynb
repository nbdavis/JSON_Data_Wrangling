```python
import requests
```

```python
response = requests.get('https://prettyprinted.com')
```

```python
#Response means that everyting was succsessful for the request. (200)
print(response.status_code)
```

```python
#Test to see if the response worked, and exists. 
if response:
    print('Response OK')
else:
    print('Response Failed')
```

```python
#Want to know the content type. It will return all of the HTML that builds the website. 
print(response.text)
```

```python
output = open('index.html','w')
output.write(response.text)
output.close()
```
